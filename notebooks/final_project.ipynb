```python
# I am completing the project in my own enviroment and not the Skills Network enviroment
# Instead of using pyfetch I'm using requests to download the dataset, which has the same result
# Download the dataset

# URL of the file
url = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-SkillsNetwork/labs/FinalModule_Coursera/data/kc_house_data_NaN.csv'

# Name for the downloaded file
filename = 'housing.csv'

# Download the file
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    # Save the content to a file
    with open(filename, 'wb') as f:
        f.write(response.content)
    print(f"File downloaded successfully as {filename}")
```

```python
# QUESTION 2
# Drop the columns "id" and "Unnamed: 0" from axis 1 using the method drop()
# Then use the method describe() to obtain a statistical summary of the data
# Make sure the inplace parameter is set to True
df.drop(['id', 'Unnamed: 0'], axis=1, inplace=True)
df.describe()
```

```python
# Split the data into training and testing sets
features =["floors", "waterfront","lat" ,"bedrooms" ,"sqft_basement" ,"view" ,"bathrooms","sqft_living15","sqft_above","grade","sqft_living"]    
X = df[features]
Y = df['price']

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.15, random_state=1)


print("number of test samples:", x_test.shape[0])
print("number of training samples:",x_train.shape[0])
```

```python
# QUESTION 9
# Create and fit a Ridge regression object using the training data
# Set the regularization parameter to 0.1
#Â Calculate the R-Squared using the test data

# Create the Ridge regression object with alpha=0.1
ridge_model = Ridge(alpha=0.1)

# Fit the model on the training data
ridge_model.fit(x_train, y_train)

# Calculate R-squared on the test set
ridge_model.score(x_test, y_test)
```

```python
# QUESTION 10
# Perform a second order polynomial transform on both the training data and testing data
# Create and fit a Ridge regression object using the training data
# Set the regularization parameter to 0.1
# Calculate the R-Squared using the test data

# Perform the second order polynomial transformation
poly = PolynomialFeatures(degree=2, include_bias=False)
x_train_poly = poly.fit_transform(x_train)
x_test_poly = poly.transform(x_test)

# Create and fit the Ridge regression object using the training data with alpha as 0.1
ridge_model = Ridge(alpha=0.1)
ridge_model.fit(x_train_poly, y_train)

# Calculate the R-Squared using the test data
ridge_model.score(x_test_poly, y_test)
```
